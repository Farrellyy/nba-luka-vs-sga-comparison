```python
import numpy as np

# Pull just this season's row for each player
season_data = combined[combined['Season'] == '2025-26']

# Stats to compare - pick ones on roughly comparable scales where possible
categories = ['PTS', 'AST', 'TRB', 'STL', 'PER']

luka_stats = season_data[season_data['Player'] == 'Luka Dončić'][categories].values.flatten().tolist()
sga_stats = season_data[season_data['Player'] == 'Shai Gilgeous-Alexander'][categories].values.flatten().tolist()

# Radar charts need the first value repeated at the end to close the loop
luka_stats += luka_stats[:1]
sga_stats += sga_stats[:1]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))


ax.plot(angles, luka_stats, linewidth=2, label='Luka Dončić', color='#552583')  # Lakers purple
ax.fill(angles, luka_stats, alpha=0.15, color='#552583')

ax.plot(angles, sga_stats, linewidth=2, label='Shai Gilgeous-Alexander', color='#EF3B24')  # Thunder orange/red
ax.fill(angles, sga_stats, alpha=0.15, color='#EF3B24')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_title('2025-26 Season Comparison', size=15, weight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))


plt.show()
```

```python
categories = ['PTS', 'AST', 'TRB', 'STL', 'PER']

luka_stats_raw = season_data[season_data['Player'] == 'Luka Dončić'][categories].values.flatten().tolist()
sga_stats_raw = season_data[season_data['Player'] == 'Shai Gilgeous-Alexander'][categories].values.flatten().tolist()

# Normalize each stat: 0 = whichever player is lower, 1 = whichever player is higher
luka_norm = []
sga_norm = []
for l, s in zip(luka_stats_raw, sga_stats_raw):
    max_val = max(l, s)
    luka_norm.append(l / max_val)
    sga_norm.append(s / max_val)

luka_norm += luka_norm[:1]
sga_norm += sga_norm[:1]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

ax.plot(angles, luka_norm, linewidth=2, label='Luka Dončić', color='#552583')
ax.fill(angles, luka_norm, alpha=0.15, color='#552583')

ax.plot(angles, sga_norm, linewidth=2, label='Shai Gilgeous-Alexander', color='#EF3B24')
ax.fill(angles, sga_norm, alpha=0.15, color='#EF3B24')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_ylim(0, 1.1)
ax.set_title('2025-26 Season Comparison (Normalized)', size=15, weight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.savefig('radar_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```

# Luka Dončić vs Shai Gilgeous-Alexander: A Statistical Comparison

A data-driven look at two of the NBA's most dynamic offensive engines — comparing 
career trajectories and current-season performance to explore the contrast between 
Luka's high-volume playmaking and SGA's elite scoring efficiency.

## Data Collection

Player statistics were pulled directly from Basketball Reference using `pandas.read_html()`, 
covering both players' full NBA careers through the 2025-26 season.

```python
url = "https://www.basketball-reference.com/players/d/doncilu01.html"
tables = pd.read_html(url)

url_sga = "https://www.basketball-reference.com/players/g/gilgesh01.html"
tables_sga = pd.read_html(url_sga)
```

## Data Cleaning
Basketball Reference pages return multiple tables 
(per-game, totals, advanced stats, etc.), so the first step is identifying 
and extracting just the tables needed.

```python
luka_per_game = tables[1]
luka_per_game

luka_advanced = tables[8]
luka_advanced


sga_per_game = tables_sga[1]
sga_per_game

sga_advanced = tables_sga[8]
sga_advanced
```

Both per-game and advanced tables include career summary rows 
(e.g. "8 Yrs") that need to be filtered out, keeping only genuine season rows.
Luka's 2024-25 season is also split into three rows due to his mid-season 
trade from Dallas to the Lakers — the combined "2TM" row is kept and the 
team-specific splits are dropped to avoid double-counting that season.

```python
luka_clean = luka_per_game[luka_per_game['Season'].str.contains('-', na=False)]
luka_clean = luka_clean[~((luka_clean['Season'] == '2024-25') & (luka_clean['Team'].isin(['DAL', 'LAL'])))]
luka_clean

luka_advanced_clean = luka_advanced[luka_advanced['Season'].str.contains('-', na=False)]
luka_advanced_clean = luka_advanced_clean[~((luka_advanced_clean['Season'] == '2024-25') & (luka_advanced_clean['Team'].isin(['DAL', 'LAL'])))]
luka_advanced_clean


sga_clean = sga_per_game[sga_per_game['Season'].str.contains('-', na=False)]
sga_clean

sga_advanced_clean = sga_advanced[sga_advanced['Season'].str.contains('-', na=False)]
sga_advanced_clean
```

The per-game and advanced stats tables are merged into a single 
dataframe per player, then both players are combined into one dataframe 
with a Player column for easy comparison.

```python
luka_full = luka_clean.merge(
    luka_advanced_clean, 
    on=['Season', 'Age', 'Team', 'Lg', 'Pos', 'G', 'GS'], 
    suffixes=('', '_adv')
)
luka_full


sga_full = sga_clean.merge(
    sga_advanced_clean, 
    on=['Season', 'Age', 'Team', 'Lg', 'Pos', 'G', 'GS'], 
    suffixes=('', '_adv')
)
sga_full


luka_full['Player'] = 'Luka Dončić'
sga_full['Player'] = 'Shai Gilgeous-Alexander'

combined = pd.concat([luka_full, sga_full], ignore_index=True)
combined
```

### Scoring Trajectory (PPG)

This chart tracks each player's points per game across their career, aligned by age 
rather than season — so we're comparing both players at equivalent career stages.

Luka has consistently scored at a higher volume than SGA at almost every matching 
age, peaking around 33-34 PPG in his mid-20s. SGA's scoring climbed more gradually 
but has accelerated sharply in the last two seasons, now matching or exceeding 
Luka's peak output.

This reflects a real difference in how their games developed — Luka was an 
elite, high-usage scorer almost immediately as a teenager, while SGA's scoring 
load grew as the Thunder built their offense increasingly around him. 

Luka was picked 3rd overall in the draft due to his EuroLeague success and was awarded rookie of the year. This helped him become the main scoring option for Dallas is his early years which explains his high points per game. 

Shai wasn't rated as high as Luka coming into the draft meaning he didn't have the offensive freedom Luka did, therefor his points per game stayed lower than Luka until a few years later in his career. 

### Shooting Efficiency (TS%)

This chart tracks True Shooting Percentage — a metric that accounts for both 
2-point, 3-point, and free throw shooting in one number — across each player's 
career by season.

SGA has held a consistently higher TS% than Luka in almost every season they've 
both played, often by a meaningful margin except for in 2021-22, SGA had a substantial decrease from the previous season going from 62% down to 56% but then bounced back to 63% in 2022-23.

Both players have trended upward over their careers, 
but SGA's efficiency has stayed elevated even as his scoring volume has grown 
alongside it.

This is the flip side of the volume story from the PPG chart: Luka generates 
more total points and shot attempts, but SGA scores more efficiently per shot 
he takes. Neither approach is "better" in isolation — they reflect different 
roles and shot profiles, with SGA relying more on getting to the free-throw 
line and high-percentage looks, while Luka creates a larger share of his own 
offense from tougher, higher-volume shots.

### Player Efficiency Rating (PER)

PER condenses a player's per-minute statistical production into a single number, 
with 15.0 representing league average — shown here as a reference line so both 
players' trajectories have context.

Both Luka and SGA have spent almost their entire careers well above league 
average, typically in the high-20s, reflecting their status as featured, 
high-usage offensive players from early in their careers. The two lines track 
each other fairly closely season to season, with neither player holding a 
dominant or consistent edge — unlike the clearer separation seen in the TS% 
chart.

In both players first 4 seasons, Luka has a much higher PER, which can be explained from his offensive freedom early in his career where he was expected to carry Dallas's offensive load. From the 2023-24 season onward is where SGA really started to shine and has exceeded Luka's PER in each season.

This is a useful check on the volume-vs-efficiency framing from the first two 
charts: PER is designed to reward both scoring volume *and* efficiency together, 
and the fact that Luka and SGA land in a similar range suggests their different 
styles — Luka's volume, SGA's efficiency — net out to comparable overall value 
on a per-minute basis, even though they get there in different ways.

### 2025-26 Season Snapshot

While the trend charts show career-long patterns, this radar chart zooms into 
the current 2025-26 season specifically — useful since both players are 
having MVP-caliber years right now, making for a clean head-to-head snapshot 
of where they each stand today.

Each stat (PTS, AST, TRB, STL, PER) is normalized so the higher of the two 
players sits at the outer edge (1.0), and the other player's value is scaled 
relative to that. This means the chart shows **relative performance per 
category**, not absolute numbers — a player sitting closer to the center on 
a given stat isn't necessarily bad at it, just lower than their opponent 
specifically. The raw numbers table below the chart fills in the actual values.

The shape shows that Luka beats SGA in every category except for PER. This shows that Luka puts up more raw numbers than SGA but on a lower efficiency.
This lines up with the season-long pattern from the 
trend charts: Luka's game leans more toward playmaking and rebounding volume, 
while SGA's strengths show up more in scoring efficiency and defensive activity.

## Conclusion

Across both career trends and this season's snapshot, Luka Dončić and Shai 
Gilgeous-Alexander represent two distinct paths to elite offensive production. 
Luka generates more total volume — higher scoring, more assists, more rebounds 
— largely by carrying a heavier offensive workload and creating shots for 
himself and others. SGA gets to a similar level of overall value (reflected in 
their comparable PER trajectories) through significantly higher efficiency, 
particularly his True Shooting Percentage, relying less on sheer volume and 
more on shot selection and getting to the free-throw line.

It's worth addressing the awards gap directly: SGA has won back-to-back MVPs 
(2024-25, 2025-26) and a Finals MVP, while Luka has yet to win any of these. On 
the surface, this could be read as the league validating SGA as the more 
valuable player. 

However, team context plays a significant role here that the individual stats 
above don't capture. SGA's MVP and championship runs coincided with the Thunder 
building a genuinely elite supporting cast around him — reinforcing their young 
core with proven veterans before their title run, and building one of the 
league's best defenses to complement his offense. Luka, by contrast, played his 
prime Dallas seasons with thinner, more inconsistent rosters around him, and 
was traded to a Lakers team still finding its footing afterward. Award voting 
tends to favor team success as much as individual production, which means it 
isn't a clean measure of who the better *player* is in isolation — it reflects 
who the better *team* is, which is a different (related but distinct) question.

This doesn't diminish what SGA has accomplished — leading a rebuild to a 
championship is a genuine achievement in its own right. But the statistical 
case above suggests the gap in hardware between the two players is wider than 
the gap in their actual on-court production.

This project reinforced a few practical data analysis lessons too: the 
importance of validating data against real-world context (catching the Lakers 
trade rows looked wrong until research confirmed they were correct), being 
careful with variable naming when working with multiple similar datasets, and 
choosing visualization scales (normalized vs raw) deliberately based on what 
story you're trying to tell.
